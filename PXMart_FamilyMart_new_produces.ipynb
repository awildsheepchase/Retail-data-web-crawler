```python
# 處理全家活動，兩張表concat後再輸出csv
# 全家用美麗湯就可以
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from datetime import date, timedelta,datetime
from fake_useragent import UserAgent
import pandas as pd
import numpy as np
from datetime import date
```

```python
# 全家活動頁：設定代理
# https://www.family.com.tw/Marketing/Event#all
user_agent = UserAgent()
faitems = requests.get("https://www.family.com.tw/Marketing/FreshFood/Product#classList0",headers={ 'user-agent': user_agent.random })

page = BeautifulSoup(faitems.text, "html.parser")
# page 查看沒有被擋下了
# 爬取活動文字
faitemlist = []
data_m = page.find_all("div",{"class":"col-md-3 col-sm-6"})

for text in data_m:
  # print(text.text)
  faitemlist.append(text.text)

faitemlist
```

```python
# 清洗
fanewitems = []
for newitem in faitemlist:
  newitem2 = newitem.strip()
  newitem3 = newitem2.split("\n")
  delet = ['']
  resitem = [i for i in newitem3 if i not in delet]
  resitem.remove("more")
  fanewitems.append(resitem)

fanewitems
```

```python
# 創建全家df
conames = ["item_name","price","kcal","description"]
newitemdf = pd.DataFrame(fanewitems,columns = conames)
newitemdf = newitemdf.drop_duplicates()
newitemdf
```

```python
# "原價＄"、"元"為不需要的資料，用切割標定符號清理
# 價格會出現如“原價$-元”無法單純用整數表示
newitemdf [["drop1","price"]]= newitemdf["price"].str.split("$",expand = True)
newitemdf[["price","drop1"]]= newitemdf["price"].str.split("元",expand = True)

newitemdf
```

```python
# 新增欄位日期
newitemdf["place_name"] = "全家"
# 創建空欄位：手動輸入商品分類用
newitemdf["item_type"] = np.nan
newitemdf["dtldate"] = date.today()
newitemdf["insertdate"] = date.today()
# 欄位排序
fa_newitems = newitemdf[["place_name", "item_name", "price","description","item_type","insertdate","dtldate"]]
# 全家df清洗完成
fa_newitems
```

```python
# 全聯比較麻煩需要selenium
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.common.action_chains import ActionChains
```

```python
# 全聯新品爬取0707
url = "https://shop.pxgo.com.tw/mweb/#/?type=hourArrive&mweb_user=tourist"
service = Service(executable_path=r'/usr/bin/chromedriver')
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome(service=service, options=options)

driver.get(url)
# 等待跳轉頁
time.sleep(3)
# 空格取代.
a = driver.find_element(By.CLASS_NAME,"q-img__container.absolute-full")
# 點擊新品頁
actions = ActionChains(driver)
actions.click(a).pause(1)
time.sleep(3)
# 獲取新品
# b = driver.find_element(By.CLASS_NAME,"left-div-list-children")
# actions.click(b).pause(1)
# time.sleep(3)
# 需要往下滑到底

# 獲取新品名稱
e = driver.find_elements(By.CLASS_NAME,"falls-font")
d = driver.find_elements(By.CLASS_NAME,"falls-active-price-number")
# 回傳list
elist = []
for value in e:
      print(value.text)
      elist.append(value.text)

dlist = []
for value in d:
      print(value.text)
      dlist.append(value.text)

driver.quit()
```

```python
print(elist)
print(dlist)
```

```python
# 創建全聯新品df
conames = ["item_name","price",]
pxnewitemdf = pd.DataFrame([elist,dlist])
# 轉
pxnewitemdf = pxnewitemdf.transpose()
pxnewitemdf.columns = conames

pxnewitemdf
```

```python
# 新增欄位 Insert column into DataFrame at specified location 人工判斷分類
pxnewitemdf.insert(2, "item_type",["女性衛生用品","餅乾","生鮮水果","生鮮水果","生鮮水果","生鮮水果","糖果","糖果","餐廚用品","小菜/滷味"], allow_duplicates=True)
# 新增欄位日期
pxnewitemdf["dtldate"] = date.today()
pxnewitemdf["insertdate"] = date.today()
pxnewitemdf["place_name"] = "pxmart"
# 全聯網站無商品描述，添加空欄位
pxnewitemdf["description"] = np.nan
pxnewitemdf
```

```python
# 欄位排序
px_newitem = pxnewitemdf[["place_name", "item_name", "price", "description","item_type","insertdate","dtldate"]]
# 全聯df清洗完成
px_newitem
```

```python
# 全聯：px_newitem,全家：fa_newitems
final = pd.concat([fa_newitems,px_newitem],ignore_index = True)
# 給id欄位名
final.index.name = "id"

final.to_csv('competitor_newitem_0707.csv', encoding = 'utf-8')
```
