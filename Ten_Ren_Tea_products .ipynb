```
import requests
from bs4 import BeautifulSoup
# 商品類別 tid 1~29中間有缺
# https://mytenren.com/shop/index.aspx?tid=3

namelist = []
pricelist = []

for i in range(30):
  r = requests.get("https://mytenren.com/shop/index.aspx",
                  params = {"tid" : i })
  p1 = BeautifulSoup(r.text, "html.parser")

  #獲取商品名稱
  data_n = p1.find_all("div",{"class":"list-title"})
  for n in data_n:
    namelist.append(n.text)

  #獲取商品價格
  data_p = p1.find_all("div",{"class":"list-ps"})
  for p in data_p:
    pricelist.append(p.text)
```

```
# 去除商品名稱前後\n
namelist = [i.strip() for i in namelist ]
namelist
```

```
# 去除商品價格前後\n
pricelist = [i.strip() for i in pricelist ]
pricelist
```

```
print("商品名稱數："+str(len(namelist)))
print("商品價格數："+str(len(pricelist)))
```

```
namedict = dict(zip(namelist, pricelist))
namedict
```

```
import pandas as pd

columns = ["name","priece"]
df = pd.DataFrame(namedict.items(), columns = columns)
df.head(20)
```
