```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By


url = "https://www.7-11.com.tw/freshfoods/hot.aspx"
options = Options()
options.add_argument("--headless")
options.add_argument("--no-sandbox")

driver = webdriver.Chrome("/usr/bin/chromedriver", options=options)
# 準備將 c 加入清單
newitems = []
driver.get(url)
c = driver.find_element(By.CLASS_NAME,"blog-list-container")
newitems.append(c.text)

driver.quit()
print(newitems)
```

```python
newitems
```

```python
# 用新上市切開
for newitem in newitems:
  newitem2 = newitem.strip()
  newitem3 = newitem2.split("新上市\n")
  delet = ['']
  resitem = [i for i in newitem3 if i not in delet]
  # resitem.remove("more")
  print(resitem)
```

```python
# 用\n切開
itemlist = []
for s in resitem:
  s2 = s.strip()
  s3 = s2.split("\n")
  # delet = ['']
  resitem = [i for i in s3]
  itemlist.append(resitem)

print(itemlist)
```

```python
# 描述內容和卡路里太亂，只取新品品名和價格
itemlist2 = []
for each in itemlist:
  eachlst = []
  eachlst.append(each[0])
  eachlst.append(each[-1])
  itemlist2.append(eachlst)
itemlist2
```

```python
import pandas as pd

conames = ["name","price"]
newitemdf = pd.DataFrame(itemlist2,columns = conames)
newitemdf = newitemdf.drop_duplicates()
newitemdf.to_csv('711newitem.csv', encoding = 'utf-8')

newitemdf
```
