```
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```
# https://www.family.com.tw/Marketing/FreshFood/Product#classList0
r = requests.get("https://www.family.com.tw/Marketing/FreshFood/Product#classList0")
                #  params = {"tid" : "3"})
p1 = BeautifulSoup(r.text, "html.parser")
# p1
test = []
data_n = p1.find_all("div",{"class":"col-md-3 col-sm-6"})
# classlist 0-2頁，共3頁內容都包含
for n in data_n:
  # print(n.text)
  test.append(n.text)
# 新品全部三頁的文字
test
```

```
newitemlis = []
for newitem in test:
  newitem2 = newitem.strip()
  newitem3 = newitem2.split("\n")
  delet = ['']
  resitem = [i for i in newitem3 if i not in delet]
  resitem.remove("more")
  newitemlis.append(resitem)

print(newitemlis)
```

```
# 要加入df 的list:newitemlis
conames = ["name","price","kcal","desc"]
newitemdf = pd.DataFrame(newitemlis,columns = conames)
newitemdf = newitemdf.drop_duplicates()
newitemdf.to_csv('familynewitem.csv', encoding = 'utf-8')
newitemdf
# 如下一個14天撈到重複的資料：concat後再 => drop_duplicates
```

```
# 活動
activitys = requests.get("https://www.family.com.tw/Marketing/Event#all")
                #  params = {"tid" : "3"})
p2 = BeautifulSoup(activitys.text, "html.parser")
# p2
test2 = []
data_m = p2.find_all("div",{"class":"card card--event"})
# classlist 0-2頁，共3頁內容都包含
for text in data_m:
  # print(text.text)
  test2.append(text.text)

test2
```

```
act_all = []
for newact in test2:
  newact2 = newact.strip()
  # 去除標籤\r
  newact2 = newact2.replace("\r","")
  newact3 = newact2.split("\n")
  delet = ['','                            ']
  resact = [i for i in newact3 if i not in delet]
  # print(resact)
  act_all.append(resact)

act_all
```

```
# 要加入df 的list:newitemlis
conames = ["class","time","name","desc"]
act_alldf = pd.DataFrame(act_all,columns = conames)
# Replace None with NaN:
act_alldf = act_alldf.fillna(value=np.nan)
# 拉掉na
act_alldf = act_alldf.dropna()
act_alldf = act_alldf.drop_duplicates()
act_alldf.to_csv('familytactivity.csv', encoding = 'utf-8')
act_alldf
# 如下一個14天撈到重複的資料：concat後再 => drop_duplicates
```
